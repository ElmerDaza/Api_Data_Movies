# Transformar los datos
* Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)

* Los valores nulos del campo rating deberán reemplazarse por el string “G” (corresponde al maturity rating: “general for all audiences”

* De haber fechas, deberán tener el formato AAAA-mm-dd

* Los campos de texto deberán estar en minúsculas, sin excepciones

* El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

# Amazon

```python
# crear columna ID
data_amazon['ID']='a'+data_amazon['show_id']
#reemplazar los valores nulos en rating
data_amazon['rating'] = data_amazon['rating'].replace(np.nan,'G')
#normalizar el formato de las fechas
data_amazon["date_added"] = pd.to_datetime(data_amazon["date_added"], format='%B %d, %Y')

#Colocar los datos strings en minusculas
# Obtener una lista de los campos de tipo string
campos_string = data_amazon.select_dtypes(include=['object']).columns.tolist()
for i in campos_string:
    #para cada campo remplazamos su contenido por minusculas
    data_amazon[i]=data_amazon[i].str.lower()

#Dividir la columna "duration" en duration_int y duration_type.
# 
#El primer grupo (\d+) busca coincidencias de uno o más dígitos (0-9) y las captura en el primer grupo.
#El segundo grupo \s*(\w+) busca 
# coincidencias de uno o más caracteres alfanuméricos y las captura en el segundo grupo, 
# permitiendo que haya cero o más espacios en blanco (\s*) entre el primer grupo y el segundo.
#  
data_amazon[['duration_int', 'duration_type']] = data_amazon['duration'].str.extract('(\d+)\s*(\w+)')

#Reemplazar "seasons" por "season":
data_amazon["duration_type"] = data_amazon["duration_type"].str.replace("seasons", "season")
```

se unifican los data frame para una mejor manipulacion desde la api

```python
#unificar los datos
full_data = pd.concat([data_amazon,data_Disney,data_hulu,data_netflix])
# cambiar los valores nulos de duration_int
full_data['duration_int']= full_data['duration_int'].replace(np.nan,'0')
#cambiar el tipo de dato para duration int
full_data['duration_int']= full_data['duration_int'].astype('int64')
full_data.dtypes
```

Columna score

```python
#Este código utiliza la biblioteca glob para encontrar 
# todos los archivos CSV en el directorio 'datasets/ratings'
#  y leerlos en un solo DataFrame. 
score = pd.concat(
    [pd.read_csv(f) for f in glob.glob('datasets/ratings/*.csv')], 
    ignore_index=True)
```

```python
# Agrupar por 'movieId' y calcular la media de 'score'
score_mean = score.groupby('movieId')['rating'].mean()
# Asignar la media correspondiente a cada fila en 'score'
full_data['score'] = score.groupby('movieId')['rating'].transform(lambda x: score_mean.loc[x.name])
#eliminar valores duplicados
full_data = full_data.drop_duplicates()
```

```python
#cambiamos las categoricas a numericas
full_data=pd.concat(
    [full_data,pd.get_dummies(full_data['type'])],axis=1)
```

```python
#eliminar la columna type
full_data = full_data.drop(columns=['type'])
full_data.head(5)
```

* analisis de sentimiento para la descripcion

```python

full_data['description']= full_data['description'].replace(np.nan,'')
```

```python

#import numpy as np

sia = SentimentIntensityAnalyzer()
# Aplicar sia.polarity_scores a toda la columna 'description'
sentimiento = full_data['description'].apply(sia.polarity_scores)
# Asignar 'negative', 'positive' o 'neutral' en función del puntaje compuesto
conditions = [    sentimiento.apply(lambda x: round(x['compound'],1)) < 0.0,
    sentimiento.apply(lambda x: round(x['compound'],1)) > 0.0,
    sentimiento.apply(lambda x: round(x['compound'],1)) == 0.0
]
choices = ['negative', 'positive', 'neutral']
full_data['feeling_description'] = np.select(conditions, choices)

```

```python
full_data = pd.concat([full_data,
                    pd.get_dummies(full_data['feeling_description'],
                                    prefix='description')],axis=1)
```

* analisis de sentimiento para el titulo

```python
full_data['title']= full_data['title'].replace(np.nan,'')
# Aplicar sia.polarity_scores a toda la columna 'description'
sentimiento = full_data['title'].apply(sia.polarity_scores)
# Asignar 'negative', 'positive' o 'neutral' en función del puntaje compuesto
conditions = [    sentimiento.apply(lambda x: round(x['compound'],1)) < 0.0,
    sentimiento.apply(lambda x: round(x['compound'],1)) > 0.0,
    sentimiento.apply(lambda x: round(x['compound'],1)) == 0.0
]
choices = ['negative', 'positive', 'neutral']
full_data['feeling_title'] = np.select(conditions, choices)
full_data = pd.concat([full_data,
                    pd.get_dummies(full_data['feeling_title'],
                                    prefix='title')],axis=1)
```
